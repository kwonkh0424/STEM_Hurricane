```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ships=pd.read_csv(r'C:\Users\kwonk\Guided Project\STEM_Hurricane\Dataset_SHIPS_RII_ATL.csv')
```

```python
# Pad the date columns with 00 for the year 2000
ships['DATE'] = ships['DATE'].apply(lambda x: str(x).zfill(6))

# Extract month from date
ships['MONTH'] = ships['DATE'].apply(lambda x: str(x)[2:4])

# Extract year from date
ships['YEAR'] = ships['DATE'].apply(lambda x: ('19' + str(x)[0:2]) if (str(x)[0:1]!= '0' and str(x)[0:1]!= '1') else ('20' + str(x)[0:2]))
ships['MONTH']=ships['MONTH'].astype(int)
ships['YEAR']=ships['YEAR'].astype(int)
ships.info()
```

```python
uni_hurr=ships['ID'].unique()

aa=pd.DataFrame()

len(uni_hurr)

        
```

```python
for hurr in uni_hurr:
    a = ships[ships['ID']==hurr]
    
    if len(a['YEAR'].unique()) == 1:
        aa=pd.concat([aa,a])
        
        
        
```
