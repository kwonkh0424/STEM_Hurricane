```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ships=pd.read_csv(r'SHIPS_ATL.csv')
```

```python
ships.head(50)
```

```python
avg_speed = ships.groupby('ID').mean()
avg_speed = avg_speed.iloc[:5]
plt.bar(avg_speed.index,avg_speed['Wind_Speed'],width=0.1)
plt.title('Average Wind Speed')
avg_speed
```
