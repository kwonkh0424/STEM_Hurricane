```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ships=pd.read_csv('https://raw.githubusercontent.com/kwonkh0424/STEM_Hurricane_Training/main/SHIPS_ATL.csv')
ships.describe()
```

```python
ships.head()
```

```python
ships.isna()
```

```python
ships = ships.replace(9999,np.NaN)
```

```python
ships.head(40)
```

```python
ships.isna()
```

```python
ships.describe()
```

```python
avg_speed = ships.groupby('YEAR').mean()
```

```python
avg_speed
```

```python
#Did the average wind speed for each year change over time?
plt.scatter(avg_speed.index,avg_speed['Wind_Speed'])
plt.title('Average Wind Speed [knots]')
#The average wind apeed does not have any pattern over time
```

```python
#Did the average DEVL24 for each year change over time? Is there any clear trend?
plt.plot(avg_speed.index,avg_speed['DELV24'])
plt.title('Average DELV24 [knots]')
#The average DELV24 does not have any pattern over time
```

```python
bins = range(1990, 2017,1)
plt.hist(ships['YEAR'],bins=bins, rwidth = 0.4)
plt.title('Average DELV24 [knots]')
```
